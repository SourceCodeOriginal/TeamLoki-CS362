```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
```

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
data.shape
```

```python
state_data = data.iloc[:, 1:].values
state_data
```

```python
state_name = data.iloc[:,0].values
state_name
```

```python
#  By using Single Link
plt.figure(figsize = (15,8))
plt.title("Using Single Link")
sch.dendrogram(sch.linkage(state_data, method='single'), labels=state_name)
plt.xticks(rotation=90)
plt.show()
```

```python
# by Using Complete Link
plt.figure(figsize = (15,8))
plt.title("Using Complete Link")
sch.dendrogram(sch.linkage(state_data, method='complete'), labels=state_name)
plt.xticks(rotation=90)
plt.show()
```

```python
# by Using Ward's distance
plt.figure(figsize = (15,8))
plt.title("Using ward's distance")
sch.dendrogram(sch.linkage(state_data, method='ward'), labels=state_name)
plt.xticks(rotation=90)
plt.show()
```

```python
# by Using group average 
plt.figure(figsize = (15,8))
plt.title("Using group average")
sch.dendrogram(sch.linkage(state_data, method='average'), labels=state_name)
plt.xticks(rotation=90)
plt.show()
```

```python
# by Using centroid method
plt.figure(figsize = (15,8))
plt.title("Using centroid method")
sch.dendrogram(sch.linkage(state_data, method='centroid'), labels=state_name)
plt.xticks(rotation=90)
plt.show()
```

```python
# by Using weighted method
plt.figure(figsize = (15,8))
plt.title("Using weighted method")
sch.dendrogram(sch.linkage(state_data, method='weighted'), labels=state_name)
plt.xticks(rotation=90)
plt.show()
```

```python
# by Using median method
plt.figure(figsize = (15,8))
plt.title("Using median method")
sch.dendrogram(sch.linkage(state_data, method='median'), labels=state_name)
plt.xticks(rotation=90)
plt.show()
```
