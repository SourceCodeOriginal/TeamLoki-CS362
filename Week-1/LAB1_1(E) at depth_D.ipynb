```python
# for 1 (E)

def in_depth(unique,current):
    subNode=[]
    child=[]
    newchild = 0
 
    
    for i in range(len(unique)-current,len(unique)):
        subNode =[]
        subNode.extend(unique[i])
        node = []
        node.extend(unique[i])
        getZeroLocation=node.index(0)+1
        
        
        if getZeroLocation+3<=9:
            
            temp=subNode[node.index(0)]
            subNode[node.index(0)]=subNode[node.index(0)+3]
            subNode[node.index(0)+3]=temp
            
            if subNode not in (unique):
                child.append(subNode)
                newchild+=1    
            subNode =[]
            subNode.extend(unique[i])

            
        if getZeroLocation-3>=1:
            
            temp=subNode[node.index(0)]
            subNode[node.index(0)]=subNode[node.index(0)-3]
            subNode[node.index(0)-3]=temp
            if subNode not in (unique):
                child.append(subNode)
                newchild+=1
            subNode =[]
            subNode.extend(unique[i])
            
           
            
        if (getZeroLocation%3)!=1:
            
            temp=subNode[node.index(0)]
            subNode[node.index(0)]=subNode[node.index(0)-1]
            subNode[node.index(0)-1]=temp
            if subNode not in (unique):
                child.append(subNode)
                newchild+=1
            subNode =[]
            subNode.extend(unique[i])
            

            
        if (getZeroLocation%3)!=0:
            
            temp=subNode[node.index(0)]
            subNode[node.index(0)]=subNode[node.index(0)+1]
            subNode[node.index(0)+1]=temp
            if subNode not in (unique):
                child.append(subNode)
                newchild+=1
            subNode =[]
            subNode.extend(unique[i])

    
    return newchild,child
            



```

```python
# for 1(E) main

import math,random,time
import numpy as np
import sys

start = np.array([2,8,3,1,6,4,7,0,5])
node =list(start)
j=0
unique = [1]
unique.append(node)
depth = int(input("Enter the depth:"))
current =1
begin = time.time()

while j<depth:
    
    newchild,child = in_depth(unique,current)
    current =newchild
    unique.extend(child)
    j+=1
    print("in depth",j,"goals are",newchild)
    

print("\n")
for i in range(len(child)):
        print(np.array(child[i]).reshape(3,3))
        print("\n")
    
end = time.time()
print("total time is:",end-begin,"seconds")

print("total numbers of node",len(unique))
total_size = sys.getsizeof(child) + sys.getsizeof(node) + sys.getsizeof(unique)
print("memory size:",total_size,"bytes")

```
