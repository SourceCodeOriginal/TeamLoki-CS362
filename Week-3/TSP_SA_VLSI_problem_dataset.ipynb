# Simulated Annealing on TSP

```python
import math,random
import numpy as np
import matplotlib.pyplot as plt


def path_cost(X,Y):
    sum = 0
    size = len(X) -1
    for i in range(0,size):
        sum +=math.sqrt((X[i]-X[i+1])**2 + (Y[i]-Y[i+1])**2)
    sum+=math.sqrt((X[0]-X[size])**2 + (Y[0]-Y[size])**2)

    return sum   

def swap_coords(X,Y,index_1,index_2):
    X[index_1],X[index_2] = X[index_2],X[index_1]
    Y[index_1],Y[index_2] = Y[index_2],Y[index_1]


def swap_random(X,Y):
    ran = len(X)-1
    pick_1 = random.randint(0,ran)
    pick_2 = random.randint(0,ran)

    X[pick_1],X[pick_2] = X[pick_2],X[pick_1]
    Y[pick_1],Y[pick_2] = Y[pick_2],Y[pick_1]

    return pick_1, pick_2


dataset = ['A131', 'B237', 'C343', 'D395','E380']

for file in dataset:
    
    print(f"Dataset is {file} \n \n")
    data = np.loadtxt(f"{file}.txt", usecols=range(1,3), dtype=np.float32)

    data =data.T
    X = data[0,:]
    Y = data[1,:]

    N = X.shape[0]

    plt.figure(figsize=(14,10))
    plt.plot(X, Y, '-o')
    plt.savefig(f'init_{file}')
    plt.title("Initial Path")
    plt.show()
    cityList =[]
    for i in range(N):
        x= round(X[i],2)
        y= round(Y[i],2)
        cityList.append([x,y])

    print("Coordinates of the cities are:\n")
    print(cityList)
    print("\nInitially root cost:",path_cost(X,Y))


    best = path_cost(X,Y)
  
    best_hist = []
    iterations = 100000
    T = 1.02
    B = 0.999

    t=0; iterations=1000; repeat=1000
    while t<repeat:
        n = 0
        while n <=iterations:

            index_1, index_2 = swap_random(X,Y)
            curr = path_cost(X,Y)
            diff = (curr - best)
            if diff < 0 :
                best = curr
                best_hist.append(best)
            elif math.exp(-(diff)/T) > random.uniform(0,1):
                best_hist.append(curr)
                T *=B
            else:
                swap_coords(X,Y,index_1,index_2)
            n += 1

        t += 1
        T *= B
        
    cityList = []
    for i in range(N):
        x= round(X[i],2)
        y= round(Y[i],2)
        cityList.append([x,y])

    print("Best route is:\n")
    print(cityList)
    
    plt.figure(figsize=(14,10))
    plt.plot(X,Y,'-o','b')
    x_values = [X[0], X[N-1]]
    y_values = [Y[0], Y[N-1]]
    plt.plot(x_values, y_values,'b')
    plt.title(f"Best Path for {file}")
    plt.savefig(f'suboptimal_{file}')
    plt.show()
    print("After applying Simulated Annealing algorithm root cost:",path_cost(X,Y))
    
    plt.figure(figsize=(14,10))
    plt.plot(best_hist)
    plt.title(f"Histogram of {file}")
    plt.savefig(f'histogram_{file}')
    plt.show()
    print("\n")
```
