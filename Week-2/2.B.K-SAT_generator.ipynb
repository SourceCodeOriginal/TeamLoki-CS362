### Write a program to randomly generate k-SAT problems. The program must acceptvalues for k, m the number of clauses in the formula, and n the number of variables. Each clause of length k must contain distinct variables or their negation. Instances generated by this algorithm belong to fixed clause length models of SAT and are known as uniform random k-SAT problems.

```python
# k-sat generator 
import random
def gen(n,m,k):
    var = []
    clauses = []
    sign=["-",""]
    # Generate variable
    while( len(var)<n):
        char = chr(random.randint(97, 122))
        if char in var:
            continue
        var.append(char)
    print("Variables are: ")
    print(var)
    i=0
    while( len(clauses)<m ):
        clause = []
        used =[]
        for j in range(k):
            a = random.choice(var)
            b = random.choice(sign)
            while (a in used): 
                a = random.choice(var)
            used.append(a)
            a = a+b
            clause.append(a)
        #print("clause generated is {} ".format (clause))

        for x in clauses:
            if(set(x)== set(clause)):
                clause= x
                
        if clause in clauses:
            continue
        clauses.append(clause) 
    #print(clauses)
    finalclause = []

    for i in range(len(clauses)):
        for j in range(len(clauses[i])):
            finalclause.append(clauses[i][j])
    #print(finalclause)
    final = ""
    i = 0
    while i < len(finalclause):
        final = final + "("
        for j in range(k):
            final = final + finalclause[i] + "|"
            i += 1
        final = final[:-1]
        final = final + ")&"
    final = final[:-1]
    print("Expression Generated is: "+final)
    #return final, var
```

```python
# driver code
n = int(input('Enter the number of variables n : '))
k = int(input('Enter the number of variables in one clause k : '))
while(k>n):
    print("wrong input!try again")
    k = int(input('Enter the number of variables in one clause: '))
m = int(input('Enter the number of clauses m (must be less than or equal to possible clause): '))
gen(n,m,k)
```
